```python
import csv
import pandas as pd

import pandas as pd
import altair as alt

bussiness_licences = "business-licences-hackathon.csv"
file = open(bussiness_licences, "r")
reader = pd.read_csv(file, sep=';')
```

```python
df_no_nan = reader.dropna(subset=['BusinessName'])
businesses_names = df_no_nan[~df_no_nan.BusinessName.str.contains('(', regex=False)].BusinessName.unique()

business_trade_names = df_no_nan[df_no_nan['BusinessName'].str.contains('(', regex=False)]["BusinessTradeName"].unique()
```

```python
dat = df_no_nan.query("BusinessName=='Tamton Networking Inc'").sort_values(by=['IssuedDate']).dropna()
dat = dat.reset_index()
dat
```

```python
chart1 = alt.Chart(dat).mark_circle().encode(
    x='LicenceNumber:N',y='IssuedDate', color='Status'
)
chart2 = alt.Chart(dat).mark_circle().encode(
    x='LicenceNumber:N',y='ExpiredDate', color='Status'
)
chart1 #add tooltip 
```
